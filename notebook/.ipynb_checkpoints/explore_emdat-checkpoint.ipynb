```python
sel_col = ['Drought', 'Earthquake', 'Volcanic activity',
       'Mass movement (dry)', 'Storm', 'Flood', 'Landslide',
       'Wildfire', 'Extreme temperature', 'Fog', 'Glacial lake outburst']
df_sub = emdat.loc[emdat['Disaster Type'].isin(sel_col)]
df_sub = df_sub.dropna(subset=['Disaster Subtype'], axis=0)
print(df_sub.shape)
```

```python
emdat.loc[emdat['Disaster Subtype'].isin(['Tropical cyclone'])]
```

```python
#pivot function
def pivot(df_in, id_col='Disaster Type', id_list=['Flood']):
    df = df_in.reset_index(drop = True)


    for id in id_list:
        #initialize columns
#         df[id+'_bin'] = 0
#         df[id+'_amt'] = 0
        df[id+'_ct'] = 0
        

#         df.loc[(df[id_col]==id), id+'_bin'] = 1
#         df.loc[(df[id_col]==id), id+'_amt'] = df["Total Damages, Adjusted ('000 US$)"].astype(float)
        df.loc[(df[id_col]==id), id+'_ct'] = 1

    return df
```

```python
# id_list= df_sub['Disaster Type'].unique()
id_list= df_sub['Disaster Subtype'].unique().tolist()
print(id_list)
df_pivot= pivot(df_sub, id_col = 'Disaster Subtype', id_list = id_list)
df_pivot
```

```python
#check pivot is doing the job 
print(emdat.loc[emdat['Disaster Subtype'].isin(['Drought'])].shape)
print(df_pivot['Drought_ct'].sum())
print(emdat.loc[emdat['Disaster Subtype'].isin(['Tropical cyclone'])].shape)
print(df_pivot['Tropical cyclone_ct'].sum())
```

```python
#aggregate columns by year
def aggregate_yrly(df):
    #one can change the year column to deal with different frequencies

   #aggregate amount
    col_amt = [col for col in df.columns if 'amt' in col]
    df_amt = df.groupby(['ISO','Year'])[col_amt].agg('sum')
    #aggregate count
    col_ct = [col for col in df.columns if '_ct' in col]
    df_ct = df.groupby(['ISO','Year'])[col_ct].agg('sum')
#     #aggregate binary
#     col_bin = [col for col in df.columns if '_bin' in col]
#     df_bin= df.groupby(['ISO','Year'])[col_bin].agg('max')

    #join
    df_out= pd.concat([df_amt, df_ct], axis=1)
#     df_out = pd.concat([df1, df_bin], axis=1)

    return df_out.reset_index()
```

```python
df_yrly = aggregate_yrly(df_pivot)
```

```python
#check aggregate_yrly is doing the job 
print(emdat.loc[emdat['Disaster Subtype'].isin(['Drought'])].shape)
print(df_yrly['Drought_ct'].sum())
print(emdat.loc[emdat['Disaster Subtype'].isin(['Tropical cyclone'])].shape)
print(df_yrly['Tropical cyclone_ct'].sum())
```

```python
#try K Means Cluster 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
#standardize amount variables 
amt_features = [col for col in df2.columns if 'amt' in col]
amt_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

preprocessor = ColumnTransformer(
    transformers=[
        ('amt', amt_transformer, amt_features)])
#process amt features 
df2[amt_features] = preprocessor.fit_transform(df2)
```

```python
df2.shape
```

```python
df2
```

```python
#sillouette score 
from sklearn.metrics import silhouette_score
sse_ = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k).fit(df2)
    sse_.append([k, silhouette_score(df2, kmeans.labels_)])
    
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1])
plt.show()
```

```python
ssd = []
for num_clusters in list(range(1,10)):
    model_clus = KMeans(n_clusters = num_clusters, max_iter=50)
    model_clus.fit(df2)
    ssd.append(model_clus.inertia_)

plt.plot(ssd)
plt.show()
```

```python
#cluster = 10 
model_clus2 = KMeans(n_clusters = 5, max_iter=100,random_state = 50)
model_clus2.fit(df2)
```
