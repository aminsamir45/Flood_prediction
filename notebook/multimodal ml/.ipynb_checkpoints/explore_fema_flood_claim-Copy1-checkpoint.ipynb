```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#https://www.fema.gov/openfema-data-page/fima-nfip-redacted-claims-v1 
df = pd.read_csv('../../data/FimaNfipClaims.csv')
#crop to after 1975 
df = df.loc[df['yearOfLoss']>=1975]
print(df.shape)
#some date analysis
print( df.dateOfLoss.min()) 
print( df.dateOfLoss.max()) 
```

### Baseline:
Using SMA and CMA for premium pricing: at a state level for now 

```python
#baseline idea: for each state use the past 10y average as insurance premium 
#https://www.geeksforgeeks.org/how-to-calculate-moving-average-in-a-pandas-dataframe/ 
#restrict to one state for now 
df1 = df.loc[df['state']=='FL']
agg1 = df1.groupby('yearOfLoss').agg({'amountPaidOnBuildingClaim':'sum'})\
.rename(columns={'amountPaidOnBuildingClaim':'amt'}) #rename column 
#simple moving average 
agg1['sma_10'] = agg1.rolling(window=10).mean()  
agg1['sma_10'] = agg1['sma_10'].shift(1)#shift down 1 to exclude current year loss  
#cumulative moving average 
agg1['cma'] = agg1['amt'].expanding().mean() 
agg1['cma'] = agg1['cma'].shift(1)#shift down 1 to exclude current year loss  
#set premium = sma 10y average 
agg1['p_sma'] = agg1['sma_10']
agg1['p_sma_cum'] = agg1['p_sma'].cumsum()
agg1['balance_sma'] = agg1['p_sma_cum']-agg1['amt']

#set premium = cma average 
agg1['p_cma'] = agg1['cma']
agg1['p_cma_cum'] = agg1['p_cma'].cumsum()
agg1['balance_cma'] = agg1['p_cma_cum']-agg1['amt']
```

```python
#plot some results 
ax = agg1[['balance_cma','balance_sma','amt']].plot(title='FL state baseline premium historical avg')
agg1[['p_cma','p_sma']].plot(ax=ax, secondary_y=True, linestyle='dotted')
```

```python
# #do this for all states
# agg = df.groupby(['state','yearOfLoss']).agg({'amountPaidOnBuildingClaim':'sum'})\
# .rename(columns={'amountPaidOnBuildingClaim':'amt'}) #rename column 

#aggregate annual loss into state level: multi-index ['state','year']
def agg_state(df):
    agg = df.groupby(['state','yearOfLoss']).agg({'amountPaidOnBuildingClaim':'sum'})\
    .rename(columns={'amountPaidOnBuildingClaim':'amt'}) #rename column 
    return agg 


#compute sma and cma for one state 
def avg_state(agg, state='FL'):
    agg1= agg.loc[state]
    #simple moving average 
    agg1['sma'] = agg1.rolling(window=10).mean()  
    agg1['sma'] = agg1['sma'].shift(1)#shift down 1 to exclude current year loss  
    #cumulative moving average 
    agg1['cma'] = agg1['amt'].expanding().mean() 
    agg1['cma'] = agg1['cma'].shift(1)#shift down 1 to exclude current year loss  
    return agg1 
    

#compute premium using average for one state 
def compute_p_avg(agg, state='FL'): 
    #compute sma and cma for one state 
    agg1 =avg_state(agg, state)
    
    #set premium = sma 10y average 
    agg1['p_sma'] = agg1['sma']
    agg1['p_sma_cum'] = agg1['p_sma'].cumsum()
    agg1['balance_sma'] = agg1['p_sma_cum']-agg1['amt']
    #set premium = cma average 
    agg1['p_cma'] = agg1['cma']
    agg1['p_cma_cum'] = agg1['p_cma'].cumsum()
    agg1['balance_cma'] = agg1['p_cma_cum']-agg1['amt']
    return agg1 

#aggregate to state level 
agg = agg_state(df)
#compute average for one state 
agg1 = compute_p_avg(agg, 'FL')
    
```

```python
#compute this for all states in a loop 
ls_states = df.state.unique().tolist()
ls_states = [x for x in ls_states if str(x)!='nan'] #drop nan value 

#aggregate to state level 
agg = agg_state(df)

states={}
for state in ls_states:
#     print(state)
    #compute average for one state 
    agg1 = compute_p_avg(agg, state)
    states[state]=agg1 
```

```python
#plot for selected states 
top7=['LA','TX','NJ','NY','FL','MS','NC']
for state in top7:
    df1 = states[state]
    plt.plot(df1['balance_cma'], label=state)
plt.legend()
plt.title('using CMA top 7 most flooded states')
```

```python
#plot for selected states 
top7=['LA','TX','NJ','NY','FL','MS','NC']
for state in top7:
    df1 = states[state]
    plt.plot(df1['balance_sma'], label=state)
plt.legend()
plt.title('using SMA')
```

```python
#number of claims by state 
bottom7=['MP','AS','GU','WY','DC','UT','ID']
for state in bottom7:
    df1 = states[state]
    plt.plot(df1['balance_cma'], label=state)
plt.legend()
plt.title('using CMA bottom 7 most flooded states')
```

```python
total_cma, total_sma = 0.0 , 0.0
for key in states:
    if key != 'MP': #remove MP for short history 
        total_cma += states[key]['balance_cma'].iloc[-1]
        total_sma += states[key]['balance_sma'].iloc[-1]

print('total cma balance for all states', total_cma) 
print('total sma balance for all states', total_sma) 
```

## Optimization idea:
prepare historical loss matrix 

```python
#prepare l_bar_ij: historical loss for location i at time j 
#aggregate annual loss into state level: multi-index ['state','year']
def agg_state(df):
    agg = df.groupby(['state','yearOfLoss']).agg({'amountPaidOnBuildingClaim':'sum'})\
    .rename(columns={'amountPaidOnBuildingClaim':'amt'}) #rename column 
    return agg 

agg = agg_state(df)
agg = agg.reset_index()
l_bar = pd.pivot_table(agg, values='amt', index=['state'],
                       columns=['yearOfLoss'], aggfunc=np.sum)
l_bar.to_csv('../../data/nfip_l_bar.csv')
```
