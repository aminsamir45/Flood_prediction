```python
# wikipedia.search("Ana E Malit")
# wikipedia.suggest("Ana E Malit")
# page = wikipedia.page('Ana E Malit')
# page.sections

def get_wiki(location):
#initiaise txt with empty string 
    txt=''
    #search similar words in wikipedia 
    for similar in wikipedia.search(location):

        search = wikipedia.WikipediaPage(similar).section('Geography')
        if search is not None:
            txt = search
            break 
    return txt 
get_wiki('Ana E Malit')
```
